## Ler um arquivo CSV com um conjunto de dados.

```python
path = '/content/drive/MyDrive/BSI/ML Health/datasets/breast_cancer.csv'

df = pd.read_csv(path)

df.head()
```

## Gráficos de distribuição de cada variável

```python
def plot_column(column: str):
  plt.figure(figsize=(16,6))
  sns.lineplot(data=df[column])
```

### `worst concave points`

```python
plot_column('worst concave points')
```

### `mean concavity`

```python
plot_column('mean concavity')
```

### `worst perimeter`

```python
plot_column('worst perimeter')
```

### `worst radius`

```python
plot_column('worst radius')
```

### `worst area`

```python
plot_column('worst area')
```

## Gráficos de dispersão dois-a-dois

```python
columns = [
  'worst concave points', 
  'mean concavity', 
  'worst perimeter', 
  'worst radius', 
  'worst area'
]
sns.pairplot(df[columns])
```

### Analisando visualmente os gráficos de dispersão, é possível verificar quais variáveis apresentam correlação dois-a-dois?

Visualmente, não consegui identificar. Entretanto, ao usar a visualização de heatmat abaixo, fica claro que existem correlações entre as variáveis `worst perimeter`, `worst radius` e `worst area`

```python
correlations = df[columns].corr()
correlations
```

```python
sns.heatmap(correlations)
plt.show()
```
